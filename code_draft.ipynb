```python
def autocorrelation(data):
    result=np.zeros((data.shape[0]))
    for m in range(data.shape[0]):
        
        for i in range(data.shape[0]):
            if (i-m) < 0 :
                result[m]=result[m]+data[i]*data[i]
            else:
                
                result[m]=result[m]+data[i]*data[i-m]
    return result
        
    
    
print(autocorrelation(data1))
```
